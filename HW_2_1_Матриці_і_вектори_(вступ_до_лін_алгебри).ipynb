3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
#1. Множення матриць А та В
result_matrix = np.dot(B, A)
print('Результат множення B*А: \n', result_matrix)
#2. Множення result_matrix на вектор С
result_matrix2 = np.dot(result_matrix, C)
print('Результат множення result_matrix*C: \n', result_matrix2)
#3. Множення result_matrix на матрицю D
result_matrix3 = np.dot(result_matrix, D)
print('Результат множення result_matrix*D: \n', result_matrix3)
```

```python
# 3. numpy.expand_dims над вектором vector
column_vector = np.expand_dims(vector, axis= 1)
print('Перетворюємо vector у стовпчик: \n',column_vector)
# За допомогою даного методу змінили структуру вектора, тепер він має 10 рядків та 1 стовпчик
print('Розмірність  column_vector: \n',column_vector.shape)
# 4. Транспонування column_vector
column_vector_t = column_vector.T
print('Транспонований column_vector: \n',column_vector_t)
print('Розмірність транспонованого column_vector: \n',column_vector_t.shape)
# Транспонування змінило структуру column_vector, тепер він має 1 рядок та 10 стовпчиків
```
