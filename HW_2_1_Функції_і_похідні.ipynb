2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
df1 = sp.sin(x)
derivative1 = sp.diff(df1, x)
df2 = sp.ln(x)
derivative2 = sp.diff(df2, x)

print(f'Похідна функції f(x) = sin(x): {derivative1}')
print(f'Похідна функції f(x) = ln(x): {derivative2}')
```

```python
# Перетворюємо формули sympy у функції
x_sin = np.linspace(0, 10, 50)
x_ln = np.linspace(0.1, 10, 50)
df1_num = sp.lambdify(x, df1, 'numpy')
derivative1_num = sp.lambdify(x, derivative1, 'numpy')
df2_num = sp.lambdify(x, df2, 'numpy')
derivative2_num = sp.lambdify(x, derivative2, 'numpy')
```

```python
# Графік sin(x) та його похідної
plt.figure(figsize=(10, 5))
plt.plot(x_sin, df1_num(x_sin), label = 'F(x)= sin(x)', color = 'green')
plt.plot(x_sin, derivative1_num(x_sin), label = "F'(x) = sin(x)", color = 'red')
plt.title('Графік sin(x) та його похідної')
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline()
plt.axvline()
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Графік f(x) = ln(x) та її похідної
plt.figure(figsize=(10, 5))
plt.plot(x_ln, df2_num(x_ln), label = 'f(x) = ln(x)', color = 'green')
plt.plot(x_ln, derivative2_num(x_ln), label = "f'(x) = ln(x)", color = 'red')
plt.title('Графік ln(x) та його похідної')
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline()
plt.axvline()
plt.grid(True)
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
df3 = x**2 - 5*x + 1
derivative3 = sp.diff(df3, x)
print('Похідна функції:', derivative3)
# Перетворення функції в numpy
df3_num = sp.lambdify(x, df3, 'numpy')
derivative3_num = sp.lambdify(x, derivative3, 'numpy')
```

```python
# Графік функції:
x_values = np.linspace(0, 100, 500)
y_df3 = df3_num(x_values)
y_derivative3 = derivative3_num(x_values)
# Будуємо графік з двома осями Y
fig, ax1 = plt.subplots(figsize=(10,6))
# Перша вісь - функція
color1 = 'tab:green'
ax1.set_xlabel('Кількість вироблених одиниць')
ax1.set_ylabel('Функція загальних витрат', color = color1)
line1 = ax1.plot(x_values, y_df3, color = color1,
                 linewidth = 2,
                 label = 'Функція витрат')
ax1.tick_params(axis = 'y', labelcolor=color1)
ax1.grid(True)

# Друга вісь - похідна функції
# Поєднуємо два різних набори даних
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Похідна функції витрат', color = color2)
line2 = ax2.plot(x_values, y_derivative3, color = color2, label = 'Витрати')
ax1.tick_params(axis = 'y', labelcolor = color2)
# Об'єднуємо легенди осей
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc = 'upper left')
plt.title('Графік функції витрат та її похідна')
fig.tight_layout()
plt.show()
```
